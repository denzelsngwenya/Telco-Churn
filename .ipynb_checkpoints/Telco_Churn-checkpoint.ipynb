## About TELCO CHURN

In this project we will analyse the Telco Churn Data. Where, Telco is a fictional telecommunications company that provides home phone and internet services. The data therefore contains:
<ol>
    <li> Customers who left the company within the last month.
    <li> Customers' about information such as gender, customer ID, Monthly and Total Charges and many more which you will see in the data description.
    <li> Customer demographics i.e age, gender, and whether or not they have dependents.
</ol>

More information about the Telco Data set can be found in [IBM](https://community.ibm.com/community/user/businessanalytics/blogs/steven-macko/2019/07/11/telco-customer-churn-1113).
